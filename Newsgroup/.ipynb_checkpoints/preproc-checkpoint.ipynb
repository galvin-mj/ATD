```python
import re
import numpy as np
import os
import random as random
import matplotlib.pyplot as plt
%matplotlib inline
from nltk.stem.porter import *
stemmer = PorterStemmer()
```

```python
# First run DataDownload.py to generate rawdocs.txt.
os.system('python DataDownload.py')

# download and read stopword list
import urllib2
request = urllib2.urlopen('http://www.textfixer.com/resources/common-english-words.txt')
output = open("stopwords.txt","w")
output.write(request.read())
output.close()

stpfile = open('stopwords.txt', 'r')
stpwrdlist = stpfile.readlines()[0].split(',')
stpfile.close()
stpwrdlist.extend(['article','writes','edu','organization'])

# set seed
np.random.seed(1415117801)
```

```python
# write both trainig and test docs in one file:
os.system('cp rawtrain.txt rawdocs.txt')
fptest = open('rawtest.txt','r')
fp = open('rawdocs.txt','a')
fp.write(fptest.read())
fp.close()
fptest.close()
```

```python
# read in training docs
docfile = open('rawdocs.txt','r')
docwrds = list()
doccnts = list()
doclbls = list() # int label for each doc
lblmap = list() # maps int labels to string labels
vocabs = list()
vocabcnts = list()
d = 0
wnum = 0
lblnum = 0
while True:
    ln = docfile.readline()
    if (len(ln) == 0):
        break
    docline = ln.split()
    doclbl = docline[0] #label
    if doclbl == 'misc.forsale':
        continue
    try:
        doclblnum = lblmap.index(doclbl)
    except ValueError:
        lblmap.append(doclbl)
        doclblnum = lblnum
        lblnum += 1
    # words
    wrdlist = list()
    cntlist = list()
    for i in range(1,len(docline)):
        wrd = docline[i]
        if wrd in stpwrdlist: # remove stop words
            continue
        wrd = str(stemmer.stem(wrd)) # stemming
        try:
            wid = vocabs.index(wrd)
            vocabcnts[wid] += 1.0
        except ValueError:
            vocabs.append(wrd)
            vocabcnts.append(1.0)
            wid = wnum
            wnum += 1
        try:
            ind = wrdlist.index(wrd)
            cntlist[ind] += 1
        except ValueError:
            wrdlist.append(wrd)
            cntlist.append(1)
    docwrds.append(wrdlist)
    doccnts.append(cntlist)
    doclbls.append(doclblnum)
    d += 1

docfile.close()
print("total number of training docs = %d" %len(docwrds))
print("total number of training words = %d" %len(vocabs))
vocabcnts = np.array(vocabcnts)
print(np.sum(vocabcnts<=0),np.sum(vocabcnts<=1),np.sum(vocabcnts<=2))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(np.log(vocabcnts))#,np.log([1,2,3,4,5,10,20]))
plt.show()
```

```python
# compute length of documents with different labels
C = len(set(lblmap))
lblNdDist = list()
for x in range(C):
	lblNdDist.append(list())
for d,dc in enumerate(doccnts):
    c = doclbls[d]
    lblNdDist[c].append(len(dc))
```

```python
# boxplt of doc lengths vs label
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(lblNdDist)
plt.show()
```

```python
# plot the same figure in log space
for c in range(C):
    lblNdDist[c] = np.log(lblNdDist[c])
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(lblNdDist)
plt.show()
```

```python
# Keeping docs with l > exp(3.5) & l < exp(5)
lblNdDist2 = []
lblsum = []
for c in range(C):
    temp = [x for x in lblNdDist[c] if x >= 3.5 and x<= 5]
    #temp = [x for x in lblNdDist[c]]
    lblNdDist2.append(np.exp(temp))
    lblsum.append(len(temp))
print(lblsum)
print('total docs = %d' %sum(lblsum))
fig = plt.figure()
ax = fig.add_subplot(211)
ax.boxplot(lblNdDist2)
ax = fig.add_subplot(212)
ax.plot(range(1,C+1),(lblsum))
ax.set_xticks(range(1,C+1))
ax.set_xticklabels(lblmap,rotation='vertical')
plt.show()
```

```python
# keeping selected training docs 
vocabs2 = list()
vocabcnts2 = list()
docwrds2 = list()
doccnts2 = list()
doclbls2 = list()
wnum = 0
for d,doc in enumerate(docwrds):
    wrdlist = list()
    cntlist = list()
    ld = np.log(len(doc))
    if ld < 3.5 or ld > 5:
        continue
    if lblmap[doclbls[d]] not in lblmap2:
        continue    
    for n,wrd in enumerate(doc):
        try:
            wid = vocabs2.index(wrd)
            vocabcnts2[wid] += doccnts[d][n]
        except ValueError:
            vocabs2.append(wrd)
            vocabcnts2.append(doccnts[d][n])
            wid = wnum
            wnum += 1
        try:
            ind = wrdlist.index(wrd)
            cntlist[ind] += doccnts[d][n]
        except ValueError:
            wrdlist.append(wrd)
            cntlist.append(doccnts[d][n])
    docwrds2.append(wrdlist)
    doccnts2.append(cntlist)
    #doclbls2.append(doclbls[d])
    lb = lblmap2.index(lblmap[doclbls[d]])
    doclbls2.append(lb)

#vocabs = vocabs2[:]
#vocabcnts = vocabcnts2[:]
print("total number of training docs = %d" %len(docwrds2))
print("total number of training words = %d" %len(vocabs2))
vocabcnts2 = np.array(vocabcnts2)
print(np.sum(vocabcnts2<=0),np.sum(vocabcnts2<=1),np.sum(vocabcnts2<=2))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(np.log(vocabcnts2))
plt.show()
```

```python
# compute length of documents with different labels
C = len(set(lblmap2))
lblNdDist = list()
for x in range(C):
	lblNdDist.append(list())
for d,dc in enumerate(doccnts2):
    c = doclbls2[d]
    lblNdDist[c].append(len(dc))
    
# boxplt of doc lengths vs label
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(lblNdDist)
plt.show()
for c in range(C):
    lblNdDist[c] = np.log(lblNdDist[c])
fig = plt.figure()
ax = fig.add_subplot(111)
ax.boxplot(lblNdDist)
plt.show()
```

```python
anomclass1 = 9 #************************
anomclass2 = 16 #************************

# training, test, and anom split:
np.random.seed(1415117801)

train_wrd = list()
train_cnt = list()
train_lbl = list()
valid_wrd = list()
valid_cnt = list()
valid_lbl = list()
test_wrd = list()
test_cnt = list()
test_lbl = list()
anom1_wrd = list()
anom1_cnt = list()
anom1_lbl = list()
anom2_wrd = list()
anom2_cnt = list()
anom2_lbl = list()

tr_id = np.arange(len(doclbls))
tranom1_id = np.where(np.array(doclbls) == anomclass1)[0]
tranom2_id = np.where(np.array(doclbls) == anomclass2)[0]
print(len(tranom1_id))
print(len(tranom2_id))
tr_id = np.setdiff1d(tr_id,tranom1_id)
tr_id = np.setdiff1d(tr_id,tranom2_id)
Dv = np.floor(0.1*len(tr_id))
Dt = np.floor(0.2*len(tr_id))
valid_id = np.random.choice(tr_id,Dv,replace=False)
tr_id = np.setdiff1d(tr_id,valid_id)
t_id = np.random.choice(tr_id,Dt,replace=False)
tr_id = np.setdiff1d(tr_id,t_id)

for d,doc in enumerate(docwrds):
    if doclbls[d] == anomclass1:
        anom1_wrd.append(doc[:])
        anom1_cnt.append(doccnts[d][:])
        anom1_lbl.append(doclbls[d])
    elif doclbls[d] == anomclass2:
        anom2_wrd.append(doc[:])
        anom2_cnt.append(doccnts[d][:])
        anom2_lbl.append(doclbls[d])
    else:
        if d in tr_id:
            train_wrd.append(doc[:])
            train_cnt.append(doccnts[d][:])
            train_lbl.append(doclbls[d])  
        elif d in valid_id:
            valid_wrd.append(doc[:])
            valid_cnt.append(doccnts[d][:])
            valid_lbl.append(doclbls[d])        
        else:
            test_wrd.append(doc[:])
            test_cnt.append(doccnts[d][:])
            test_lbl.append(doclbls[d])         
    
print('anomalous docs = %d' %len(anom1_wrd))
print('anomalous docs = %d' %len(anom2_wrd))
print('validation docs = %d' %len(valid_wrd))
print('#train = %d, #normal test = %d' %(len(train_wrd),len(test_wrd)))
```

```python
os.system('mkdir -p data')

# write words
wrdmap = {}
wnum = 0
fp = open('data/vocabs.txt','w+')
for d,doc in enumerate(train_wrd):
    for n,w in enumerate(doc):
        if w not in wrdmap:
            fp.write(w+'\n')
            wrdmap.update({w:wnum})
            wnum += 1
fp.close()

def write_docs(wrds, cnts, wrdmap):
    widlist = list()
    cntlist = list()
    for n,w in enumerate(wrds):
        if w in wrdmap:
            wid = wrdmap[w]
        else:
            continue
        try:
            ind = widlist.index(wid)
            cntlist[ind] += cnts[n]
        except ValueError:
            widlist.append(wid)
            cntlist.append(cnts[n])
    ld = len(widlist)
    doctxt = str(ld)+ ' '
    for n,w in enumerate(widlist):
        doctxt += str(w) + ':' + str(cntlist[n]) + ' '
    return doctxt


# training docs
fp1 = open('data/trdocs.txt','w+')
fp2 = open('data/trlbls.txt','w+')
for d,doc in enumerate(train_wrd):
    lbl = train_lbl[d]
    doctxt = write_docs(train_wrd[d], train_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()


# validation set
fp1 = open('data/vdocs.txt','w+')
fp2 = open('data/vlbls.txt','w+')
for d,doc in enumerate(valid_wrd):
    lbl = valid_lbl[d]
    doctxt = write_docs(valid_wrd[d], valid_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()

# clean test set
fp1 = open('data/ntdocs.txt','w+')
fp2 = open('data/ntlbls.txt','w+')
for d,doc in enumerate(test_wrd):
    lbl = test_lbl[d]
    doctxt = write_docs(test_wrd[d], test_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()

# test set + 10% of class 7 docs
fp1 = open('data/tdocs10.txt','w+')
fp2 = open('data/tlbls10.txt','w+')
for d,doc in enumerate(test_wrd):
    lbl = test_lbl[d]
    doctxt = write_docs(test_wrd[d], test_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
Danom = len(anom1_wrd)
dlist = np.random.choice(np.arange(Danom), np.floor(Danom*0.1), replace = False)
for d,doc in enumerate(anom1_wrd):
    if d not in dlist:
        continue
    lbl = anom1_lbl[d]
    doctxt = write_docs(anom1_wrd[d], anom1_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
Danom = len(anom2_wrd)
dlist = np.random.choice(np.arange(Danom), np.floor(Danom*0.1), replace = False)
for d,doc in enumerate(anom2_wrd):
    if d not in dlist:
        continue
    lbl = anom2_lbl[d]
    doctxt = write_docs(anom2_wrd[d], anom2_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()

# test set + 20% of class 7 docs
fp1 = open('data/tdocs20.txt','w+')
fp2 = open('data/tlbls20.txt','w+')
for d,doc in enumerate(test_wrd):
    lbl = test_lbl[d]
    doctxt = write_docs(test_wrd[d], test_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
Danom = len(anom1_wrd)
dlist = np.random.choice(np.arange(Danom), np.floor(Danom*0.2), replace = False)
for d,doc in enumerate(anom1_wrd):
    if d not in dlist:
        continue
    lbl = anom1_lbl[d]
    doctxt = write_docs(anom1_wrd[d], anom1_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
Danom = len(anom2_wrd)
dlist = np.random.choice(np.arange(Danom), np.floor(Danom*0.2), replace = False)
for d,doc in enumerate(anom2_wrd):
    if d not in dlist:
        continue
    lbl = anom2_lbl[d]
    doctxt = write_docs(anom2_wrd[d], anom2_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()
```
