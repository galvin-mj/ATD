```python
import re
import numpy as np
import os
import random as random
import matplotlib.pyplot as plt
from nltk.stem.porter import *
%matplotlib inline
```

```python
# First run DataDownload.py to generate rawdocs.txt.
os.system('python DataDownload.py')

# download and read stopword list
import urllib2
request = urllib2.urlopen('http://www.textfixer.com/resources/common-english-words.txt')
output = open("stopwords.txt","w")
output.write(request.read())
output.close()

stpfile = open('stopwords.txt', 'r')
stpwrdlist = stpfile.readlines()[0].split(',')
stpfile.close()

# set seed
np.random.seed(1676507445)
```

```python
# read labels
fp = open('all-topics-strings.txt','r')
temp = fp.readlines()
fp.close()
all_lbls = [x.split()[0] for x in temp]
print('Total number of labels = %d' %len(all_lbls))
```

```python
# set up stemmer
stemmer = PorterStemmer()
```

```python
# read the text file

docfile = open('rawdocs.txt','r')
lblcnt = np.zeros(len(all_lbls)) # number of docs with each label
docwrds = list()
doccnts = list()
doclbls = list() # int label for each doc
lblmap = list() # maps int labels to string labels
vocabs = list()
vocabcnts = list()
docinds = list()
d = 0
wnum = 0
lblnum = 0
while True:
	docline = docfile.readline()
	if (len(docline) == 0):
		break
	docline = docline.split()
	doclbl = docline[0]
	X = doclbl.split('_')
	lX = float(len(X))
	for x in X:
		ind = all_lbls.index(x)
		lblcnt[ind] += 1.0/lX
	try:
		doclblnum = lblmap.index(doclbl)
	except ValueError:
		lblmap.append(doclbl)
		doclblnum = lblnum
		lblnum += 1
	doclbls.append(doclblnum)
    
    # words
	wrdlist = list()
	cntlist = list()
	for i in range(1,len(docline)):
		wrd0 = docline[i]
		if wrd0 in stpwrdlist: # skip stpwrds
			continue
		if len(wrd0) <= 2: # skip short wrds
			continue
		wrd = str(stemmer.stem(wrd0))
		try:
			wid = vocabs.index(wrd)
			vocabcnts[wid] += 1
		except ValueError:
			vocabs.append(wrd)
			vocabcnts.append(1)
			wid = wnum
			wnum += 1
		try:
			ind = wrdlist.index(wrd)
			cntlist[ind] += 1
		except ValueError:
			wrdlist.append(wrd)
			cntlist.append(1)
	docwrds.append(wrdlist)
	doccnts.append(cntlist) 
	d += 1
	if d%5000 == 0:
		print(d)

docfile.close()
print("total number of training docs = %d" %len(docwrds))
print("total number of training words = %d" %len(vocabs))
vocabcnts = np.array(vocabcnts)
print(np.sum(vocabcnts<=0),np.sum(vocabcnts<=1),np.sum(vocabcnts<=2))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(np.log(vocabcnts))#,np.log([1,2,3,4,5,10,20]))
plt.show()
```

```python
# number of docs with each label

fig = plt.figure()
ax = fig.add_subplot(221)
ax.hist(lblcnt)
ax = fig.add_subplot(222)
ax.plot(lblcnt)
ax = fig.add_subplot(223)
ax.hist(np.log(lblcnt+1e-2))
ax = fig.add_subplot(224)
ax.plot(np.log(lblcnt+1e-2))
plt.show()
print(np.median(lblcnt))
```

```python
# histogram of length of documents

docld = np.zeros(len(docwrds))
docnd = np.zeros(len(docwrds))
for d,cnt in enumerate(doccnts):
    docld[d] = (len(cnt))
    docnd[d] = (np.sum(cnt))

fig = plt.figure()
ax = fig.add_subplot(221)
ax.hist(docld)
ax = fig.add_subplot(222)
ax.hist(np.log(docld))
ax = fig.add_subplot(223)
ax.hist(docnd)
ax = fig.add_subplot(224)
ax.hist(np.log(docnd))
plt.show()
```

```python
# keeping selected training docs 
vocabs2 = list()
vocabcnts2 = list()
docwrds2 = list()
doccnts2 = list()
doclbls2 = list()
wnum = 0
for d,doc in enumerate(docwrds):
    wrdlist = list()
    cntlist = list()
    ld = np.log(len(doc))
    if ld <= 3.5 or ld >= 5:
        continue
    #print(lblmap[doclbls[d]])
    X = lblmap[doclbls[d]].split('_')
    
    skipdoc = 0
    for x in X:
        ind = all_lbls.index(x)
        if np.log(lblcnt[ind]+1e-2) < 3.5:
            skipdoc = 1
            break
    if skipdoc == 1:
        continue
    if 'livestock' in lblmap[doclbls[d]]:
        continue
    if 'acq' in lblmap[doclbls[d]]:
        continue
    
    wrdlist = doc #***************** Remove these 2 lines if any word may be deleted
    cntlist = doccnts[d]  #*****************
    for n,wrd in enumerate(doc):
        try:
            wid = vocabs2.index(wrd)
            vocabcnts2[wid] += doccnts[d][n]
        except ValueError:
            vocabs2.append(wrd)
            vocabcnts2.append(doccnts[d][n])
            wid = wnum
            wnum += 1
        #try:
        #    ind = wrdlist.index(wrd)
        #    cntlist[ind] += doccnts[d][n]
        #except ValueError:
        #    wrdlist.append(wrd)
        #    cntlist.append(doccnts[d][n])
    docwrds2.append(wrdlist)
    doccnts2.append(cntlist)
    #doclbls2.append(doclbls[d])
    lb = doclbls[d]
    doclbls2.append(lb)


print("total number of training docs = %d" %len(docwrds2))
print("total number of training words = %d" %len(vocabs2))
vocabcnts2 = np.array(vocabcnts2)
print(np.sum(vocabcnts2<=0),np.sum(vocabcnts2<=1),np.sum(vocabcnts2<=2))
fig = plt.figure()
ax = fig.add_subplot(111)
ax.hist(np.log(vocabcnts2))
plt.show()
```

```python
docwrds = docwrds2[:]
doccnts = doccnts2[:]
doclbls = doclbls2[:]
vocabs = vocabs2[:]
vocabcnts = vocabcnts2.copy()
```

```python
print('number of unique labels %d' %len(lblmap))
# number of docs with >1 label
ml = 0
for doclbl in doclbls:
    if '_' in lblmap[doclbl]:
        ml += 1
print('docs with more than 1 lbl %d' %ml)
anomcnt = 0
for x in lblmap:
	if 'ship' in x:
		anomcnt += 1    
print('numbe of lbls with ship lbl = %d' %anomcnt)
ml = 0
for doclbl in doclbls:
    if 'ship' in(lblmap[doclbl]):
        ml += 1
print('# docs with ship lbls = %d' %ml)
ml = 0
for doclbl in doclbls:
    if ('coffee' in (lblmap[doclbl])):
        ml += 1
print('# docs with coffee lbls = %d' %ml)
ml = 0
for doclbl in doclbls:
    if ('coffee' in(lblmap[doclbl])) and ('ship' in(lblmap[doclbl])):
        ml += 1
print('# docs with coffee and ship lbls = %d' %ml)
```

```python
# separate docs that have 'ship' and 'coffee' in their lbls
anom1_id = []
anom2_id = []
remove_id = []
for d, lbl in enumerate(doclbls):
    if ('ship' in (lblmap[lbl])) and ('coffee' not in lblmap[lbl]):
        anom1_id.append(d)
    if ('ship' not in (lblmap[lbl])) and ('coffee' in lblmap[lbl]):
        anom2_id.append(d)
    if ('ship' in (lblmap[lbl])) and ('coffee' in lblmap[lbl]):
        remove_id.append(d)
remove_id = np.array(remove_id)
anom_id = anom1_id[:]
anom_id.extend(anom2_id)
anom_id = np.array(anom_id)
anom1_id = np.array(anom1_id)
anom2_id = np.array(anom2_id)
print('# anomalous docs = %d, %d --> %d' %(len(anom1_id),len(anom2_id), len(anom_id)))
print(len(remove_id))
```

```python
np.random.seed(1000000001)#1676507445)
# training, test, and anom split:
train_wrd = list()
train_cnt = list()
train_lbl = list()
valid_wrd = list()
valid_cnt = list()
valid_lbl = list()
test_wrd = list()
test_cnt = list()
test_lbl = list()
anom_wrd = list()
anom_cnt = list()
anom_lbl = list()

all_id = np.arange(len(doclbls))
all_id = np.setdiff1d(all_id,anom_id)
all_id = np.setdiff1d(all_id,remove_id) ####### remove docs with both 'ship' and 'coffee' lbls (2 docs)

DD = len(all_id)
tr_id = np.random.choice(all_id,np.floor(0.75*DD),replace=False)
all_id = np.setdiff1d(all_id,tr_id)
valid_id = np.random.choice(all_id,np.floor(0.05*DD),replace=False)
test_id = np.setdiff1d(all_id,valid_id)

for d,doc in enumerate(docwrds):

    if d in tr_id:
        train_wrd.append(doc)
        train_cnt.append(doccnts[d])
        train_lbl.append(doclbls[d])        
    elif d in test_id:
        test_wrd.append(doc)
        test_cnt.append(doccnts[d])
        test_lbl.append(doclbls[d])    
    elif d in valid_id:
        valid_wrd.append(doc)
        valid_cnt.append(doccnts[d])
        valid_lbl.append(doclbls[d])        
    elif d in anom_id:
        anom_wrd.append(doc)
        anom_cnt.append(doccnts[d])
        anom_lbl.append(doclbls[d])
    elif d in remove_id:
        print ('skip doc')
    else:
        print('logic error')
        
    
print('anomalous docs = %d' %len(anom_wrd))
print('validation docs = %d' %len(valid_wrd))
print('#train = %d, #normal test = %d' %(len(train_wrd),len(test_wrd)))
```

```python
os.system('mkdir -p data')

# write words
wrdmap = {}
wnum = 0
fp = open('data/vocabs.txt','w+')
for d,doc in enumerate(train_wrd):
    for n,w in enumerate(doc):
        if w not in wrdmap:
            fp.write(w+'\n')
            wrdmap.update({w:wnum})
            wnum += 1
fp.close()

def write_docs(wrds, cnts, wrdmap):
    widlist = list()
    cntlist = list()
    for n,w in enumerate(wrds):
        if w in wrdmap:
            wid = wrdmap[w]
        else:
            continue
        try:
            ind = widlist.index(wid)
            cntlist[ind] += cnts[n]
        except ValueError:
            widlist.append(wid)
            cntlist.append(cnts[n])
    ld = len(widlist)
    doctxt = str(ld)+ ' '
    for n,w in enumerate(widlist):
        doctxt += str(w) + ':' + str(cntlist[n]) + ' '
    return doctxt


# training docs
fp1 = open('data/trdocs.txt','w+')
fp2 = open('data/trlbls.txt','w+')
for d,doc in enumerate(train_wrd):
    lbl = train_lbl[d]
    doctxt = write_docs(train_wrd[d], train_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()


# validation set
fp1 = open('data/vdocs.txt','w+')
fp2 = open('data/vlbls.txt','w+')
for d,doc in enumerate(valid_wrd):
    lbl = valid_lbl[d]
    doctxt = write_docs(valid_wrd[d], valid_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()

# clean test set
fp1 = open('data/ntdocs.txt','w+')
fp2 = open('data/ntlbls.txt','w+')
for d,doc in enumerate(test_wrd):
    lbl = test_lbl[d]
    doctxt = write_docs(test_wrd[d], test_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()

# test set + 10% of anomalous classes
fp1 = open('data/tdocs10.txt','w+')
fp2 = open('data/tlbls10.txt','w+')
for d,doc in enumerate(test_wrd):
    lbl = test_lbl[d]
    doctxt = write_docs(test_wrd[d], test_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
Danom = len(anom_wrd)
dlist = np.random.choice(np.arange(Danom), np.floor(Danom*0.1), replace = False)
for d,doc in enumerate(anom_wrd):
    if d not in dlist:
        continue
    lbl = anom_lbl[d]
    doctxt = write_docs(anom_wrd[d], anom_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()

# test set + 20% of anomalous classes
fp1 = open('data/tdocs20.txt','w+')
fp2 = open('data/tlbls20.txt','w+')
for d,doc in enumerate(test_wrd):
    lbl = test_lbl[d]
    doctxt = write_docs(test_wrd[d], test_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
Danom = len(anom_wrd)
dlist = np.random.choice(np.arange(Danom), np.floor(Danom*0.2), replace = False)
for d,doc in enumerate(anom_wrd):
    if d not in dlist:
        continue
    lbl = anom_lbl[d]
    doctxt = write_docs(anom_wrd[d], anom_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()


# test set + 50% of class 7 docs
fp1 = open('data/tdocs50.txt','w+')
fp2 = open('data/tlbls50.txt','w+')
for d,doc in enumerate(test_wrd):
    lbl = test_lbl[d]
    doctxt = write_docs(test_wrd[d], test_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')

#write 25% of docs with 'ship' lbl and all of docs with 'coffee' lbl
anom_ship = list()
anom_coffee = list()
for d,lbl in enumerate(anom_lbl):
    if 'ship' in lblmap[lbl]:
        anom_ship.append(d)
    if 'coffee' in lblmap[lbl]:
        anom_coffee.append(d)
anom_ship = np.array(anom_ship)
anom_coffee = np.array(anom_coffee)
dlist = np.random.choice(anom_ship, np.floor(len(anom_ship)*.25), replace = False)
for d,doc in enumerate(anom_wrd):
    if d not in dlist:
        continue
    lbl = anom_lbl[d]
    doctxt = write_docs(anom_wrd[d], anom_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
dlist = np.random.choice(anom_coffee, np.floor(len(anom_coffee)*1), replace = False)
for d,doc in enumerate(anom_wrd):
    if d not in dlist:#anom_coffee:
        continue
    lbl = anom_lbl[d]
    doctxt = write_docs(anom_wrd[d], anom_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')    
fp1.close()
fp2.close()

# test set + all of class 7 docs
fp1 = open('data/tdocs100.txt','w+')
fp2 = open('data/tlbls100.txt','w+')
for d,doc in enumerate(test_wrd):
    lbl = test_lbl[d]
    doctxt = write_docs(test_wrd[d], test_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
Danom = len(anom_wrd)
dlist = np.random.choice(np.arange(Danom), np.floor(Danom*1), replace = False)
for d,doc in enumerate(anom_wrd):
    if d not in dlist:
        continue
    lbl = anom_lbl[d]
    doctxt = write_docs(anom_wrd[d], anom_cnt[d], wrdmap)
    fp1.write(doctxt + '\n')
    fp2.write(lblmap[lbl]+'\n')
fp1.close()
fp2.close()
```
